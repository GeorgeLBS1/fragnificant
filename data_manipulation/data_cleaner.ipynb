# Manipulating the data

## Read json file

```python
# Path: data/fragances.json
# Opening JSON file
f = open("../data/fragances.json", encoding="utf8")

# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
for i in range(len(data)):
    data[i] = flatten_json(data[i])
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('../data/fragances.csv', index=False)
```
