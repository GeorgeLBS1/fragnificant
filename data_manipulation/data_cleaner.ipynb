```python
#!/usr/bin/env python3
import pandas as pd
import json
from utils.normalizer import Normalizer
```

# Manipulating the data

## Read json file

```python
# Path: data/fragances.json
# Opening JSON file
f = open("../data/fragances.json", encoding="utf8")

# returns JSON object as 
# a dictionary
data = json.load(f)

# Closing file
f.close()
```

## Get all distinct values from fragances attributes

```python
nml = Normalizer()
main_accords = nml.get_all_main_accords(data)
top_notes = nml.get_all_top_notes(data)
middle_notes = nml.get_all_middle_notes(data)
base_notes = nml.get_all_base_notes(data)
```

```python
print("All Main accords: ", len(main_accords))
print("All Top notes: ", len(top_notes))
print("All Middle notes: ", len(middle_notes))
print("All Base notes: ", len(base_notes))
```

```python
print(top_notes)
```

# Normalize the data

```python
ftd = nml.flatten_data(data, main_accords, top_notes, middle_notes, base_notes)
```

```python
print("Flattened data: ")
print(ftd[1])
```

# Save the data

```python
df = pd.DataFrame(ftd)
```

```python
print("Dataframe: ")
print(df.shape)
```
